```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from IPython.display import display
```

```python
df = pd.read_csv("/content/drive/MyDrive/data/imdb_top_movies_1980_2026.csv")
```

```python
df['genres'] = df['genres'].fillna('')
tfidf = TfidfVectorizer(stop_words='english')
genre_matrix = tfidf.fit_transform(df['genres'])
cosine_sim = cosine_similarity(genre_matrix, genre_matrix)
indices = pd.Series(df.index, index=df['title'].str.lower()).drop_duplicates()
def recommend(movie_title, top_n=10):
    movie_title = movie_title.lower()
    if movie_title not in indices:
        return "Movie not found."
    idx = indices[movie_title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_df = df.iloc[[i[0] for i in sim_scores]].copy()
    sim_df['similarity_score'] = [i[1] for i in sim_scores]
    candidates = sim_df.sort_values(by='similarity_score', ascending=False).iloc[1:51]
    recommendations = candidates.sort_values(by='average_rating', ascending=False).head(top_n)
    return recommendations[['title', 'genres', 'average_rating', 'num_votes']]
movie_name = input("Enter a movie name: ")
recommendations = recommend(movie_name)
if isinstance(recommendations, pd.DataFrame):
    display(recommendations.reset_index(drop=True).style.hide(axis='index'))
else:
    print(recommendations)
```
